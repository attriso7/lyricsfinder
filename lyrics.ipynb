Searching song on Genius API with the help of Lyricsgenius library

```python
from requests.exceptions import HTTPError, Timeout
from lyricsgenius import Genius
from api_key import *
while True:
    try:
        genius = Genius(api_key, remove_section_headers=True)
        til = input("Song Title > ")
        art = input("Artist Name > ")
        song = genius.search_song(til, art)
        print('\n', song.lyrics)
        print(type(song))
    except HTTPError as e:
        print(e.errno)
        print(e.args[0])
        print(e.args[1])
    except Timeout:
        pass
    print("Again? (y/n)")
    again = input("> ")
    if again == "n": break
```
